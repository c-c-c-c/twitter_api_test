```python
# path = './mylog0701_0704timepast.log'
path = "./0701expt/0630_2157/mylog.log" # 最新のもの

f = open(path)

plist =[]
while True:
    data = f.readline()
    if data == '':
        break
    
    if "perplexity" in data: 
        print (data)
        plist.append(data)

f.close()
```
