```python
# path = './mylog0701_0704timepast.log'
path = "../experiment/0803expt/mylog.log" # 最新のもの

f = open(path)

plist =[]
all_list = []
while True:
    data = f.readline()
    if data == '':
        break
    
    all_list.append(data)
    
    if "perplexity" in data: 
        print (data)
        plist.append(data)

f.close()
```
