```python
from datetime import datetime
import datetime
d_today = datetime.datetime.now()- datetime.timedelta(days=20)
d_today = pytz.utc.localize(d_today).astimezone(pytz.timezone("Asia/Tokyo"))

since = dt_to_isostr(d_today)
```

```python
import requests
import os
import json
import Bearer

# To set your environment variables in your terminal run the following line:
# export 'BEARER_TOKEN'='<your_bearer_token>'
bearer_token = Bearer.BEARER

search_url = "https://api.twitter.com/2/tweets/search/all"

# Optional params: start_time,end_time,since_id,until_id,max_results,next_token,
# expansions,tweet.fields,media.fields,poll.fields,place.fields,user.fields
query_params = {'query': '#SHOW',
                'tweet.fields': 'author_id',
                'max_results' : 100,
                "start_time":since }#, "until":until}


def create_headers(bearer_token):
    headers = {"Authorization": "Bearer {}".format(bearer_token)}
    return headers


def connect_to_endpoint(url, headers, params):
    response = requests.request("GET", search_url, headers=headers, params=params)
    print(response.status_code)
    if response.status_code != 200:
        raise Exception(response.status_code, response.text)
    return response.json()


def main():
    headers = create_headers(bearer_token)
    json_response = connect_to_endpoint(search_url, headers, query_params)
    print(json.dumps(json_response, indent=4, sort_keys=True))


if __name__ == "__main__":
    main()
```

```python
from datetime import datetime
def iso_to_jstdt(iso_str):
    dt = None
    try:
        dt = datetime.strptime(iso_str, '%Y-%m-%dT%H:%M:%S.%fZ')
        dt = pytz.utc.localize(dt).astimezone(pytz.timezone("Asia/Tokyo"))
    except ValueError:
        try:
            dt = datetime.strptime(iso_str, '%Y-%m-%dT%H:%M:%S.%f%z')
            dt = dt.astimezone(pytz.timezone("Asia/Tokyo"))
        except ValueError:
            pass
    return dt
```

```python
def dt_to_isostr(dt):
    if dt is None:
        return ''
    return dt.isoformat()
```

```python
def dt_to_str(dt):
    if dt is None:
        return ''
    return dt.strftime('%Y/%m/%d %H:%M:%S')
```
